```python
##2016 Data

input_directory = 'GEE_Dynamic_2016_three_class_np_label/'
output_directory = 'GEE_Dynamic_2016_3_class_value_image/'

for fn in os.listdir(input_directory):
        if fn.endswith("npy"):
          #print(fn)
          im = np.load(input_directory + fn)
          #print(input_directory + fn)
          Image.fromarray(im[:,:,0]).save(output_directory + fn.split(".")[0] + '.png')
          #print(output_directory + fn.split(".")[0] + '.png' + '......... completed')
          print(fn.split(".")[0] + '......... completed')
```
